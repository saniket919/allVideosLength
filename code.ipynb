```python
import os
from moviepy.video.io.VideoFileClip import VideoFileClip

def print_video_lengths(folder_path="."):
    """
    Prints the length of all videos in a folder in minutes and hours to the console,
    as well as the total length of all videos.

    Args:
        folder_path (str): The path to the folder. Defaults to the current working directory.
    """
    total_hours = 0
    total_minutes = 0
    found_videos = False
    for filename in os.listdir(folder_path):
        if filename.endswith(".mp4") or filename.endswith(".avi") or filename.endswith(".mov"):
            found_videos = True
            file_path = os.path.join(folder_path, filename)
            clip = VideoFileClip(file_path)
            duration = clip.duration
            hours, rem = divmod(duration, 3600)
            minutes, seconds = divmod(rem, 60)
            print(f"{filename}: {int(hours)} hours, {int(minutes)} minutes")
            total_hours += hours
            total_minutes += minutes
    if not found_videos:
        print("No video files found in specified folder.")
    else:
        total_minutes += total_hours * 60
        print(f"Total length: {total_hours} hours, {total_minutes} minutes")
print_video_lengths()
```
